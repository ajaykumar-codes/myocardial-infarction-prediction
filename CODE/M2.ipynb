# Data visualization:

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('heart.csv')
```

```python
df = data.dropna()
```

```python
df.columns
```

```python
sns.countplot(x="Mortality", data=df, palette="husl")
```

```python
df['Age'].hist(figsize=(5,5), color='brown', alpha=1)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age and its Count')
```

```python
df['chol'].hist(figsize=(5,5), color='brown', alpha=1)
plt.xlabel('Chol')
plt.ylabel('Count')
plt.title('Age and its Count')
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'Smoking')
```

```python
piechart(df, 'Gender')
```

```python
g = sns.catplot("Mortality", data=df, kind='count', hue='Age', margin_titles=False)
g.set_ylabels('Number of persons')
g.fig.set_figheight(5)
g.fig.set_figwidth(20)
```

```python
sns.pairplot(df)
```

```python
fig, ax = plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot = True, ax=ax)
plt.show()
```
